```python
import pandas as pd
import glob
```

```python
path = r'ercot_data/' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    ercotData_df = pd.read_csv(filename, index_col=None, header=0)
    li.append(ercotData_df)

frame = pd.concat(li, axis=0, ignore_index=True)
```
