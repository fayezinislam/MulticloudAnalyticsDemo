```python
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Stations, Daily
```

```python
# Get weather stations ordered by distance to Vancouver, BC
stations = Stations()
# Fetch closest station (limit = 1)
stations = stations.nearby(31.9686 , -99.9018)
stations = stations.region('US','TX')
station = stations.fetch(1000)
station
```

```python
# Get daily data for 2018 at the selected weather station
data = Daily(station, start = datetime(2010, 1, 1), end = datetime(2022, 2, 28))
# Fetch Pandas DataFrame
data = data.fetch()
```

```python
# Plot line chart including average, minimum and maximum temperature
data.plot(y = ['tavg', 'tmin', 'tmax'], kind = 'line')
# Show chart
plt.show()
```

```python
data.to_csv('weather_data/texas_weather_historical.csv')
```
